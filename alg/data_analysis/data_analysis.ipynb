```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('training_performance_adjusted_data.csv')
```

```python
df.head()
```

```python
df.type5_easy.plot(kind='box')
```

```python
distortion = []
K = range(1, 15)
for k in K:
    model = KMeans(n_clusters=k, random_state=12345)
    model.fit(df)
    distortion.append(model.inertia_)
    
plt.figure(figsize=(12, 8))
plt.plot(K, distortion, 'bx-')
plt.xlabel(' ')
plt.ylabel('  ')
plt.show()
```
